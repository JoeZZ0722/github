```python
# 16年3月到9月的数据清洗合成
df1 = pd.DataFrame(pd.read_excel('16.39.xlsx',sep = '|',encoding='GBK'))
df1.rename(columns={u'时间':'Date',u'地市':'Place',u'流量（Gb）':'DataflowByDay_4G_GB'},inplace = True)
df1 = df1.groupby(df1['Place'])
for name,groupData in df1:
    if name.encode('utf-8') == '镇江':
        f1 = pd.DataFrame(groupData)
        f1.drop('Place',axis = 1, inplace = True)
        f1['time'] = f1['Date'].apply(lambda x:str(x.encode('utf-8')).split()[0])
        f1.index = pd.to_datetime(f1['time'])
        f1.index.rename('Date',inplace = True)
        f1.drop({'Date','time'},axis = 1, inplace = True)
f1
```

```python
# 16年9月到12月的数据清洗合成
df2 = pd.DataFrame(pd.read_csv('2016.9.csv',sep = ',',encoding='GBK'))
df2.rename(columns={u'时间':'Date',u'地市':'Place',u'流量（Gb）':'DataflowByDay_4G_GB'},inplace = True)
df2 = df2.groupby(df2['Place'])
for name,groupData in df2:
    if name.encode('utf-8') == '镇江':
        f2 = pd.DataFrame(groupData)
        f2.rename(columns={u'时间':'Date',u'流量（GB）':'DataflowByDay_4G_GB'},inplace = True)
        f2.drop('Place',axis = 1, inplace = True)
        f2 = f2.set_index(pd.to_datetime(f2['Date']))
        f2.drop('Date',axis = 1, inplace = True)

df3 = pd.DataFrame(pd.read_csv('2016.10.csv',sep = ',',encoding='GBK'))
df3.rename(columns={u'时间':'Date',u'地市':'Place',u'流量（Gb）':'DataflowByDay_4G_GB'},inplace = True)
df3 = df3.groupby(df3['Place'])
for name,groupData in df3:
    if name.encode('utf-8') == '镇江':
        f3 = pd.DataFrame(groupData)
        f3.rename(columns={u'时间':'Date',u'流量（GB）':'DataflowByDay_4G_GB'},inplace = True)
        f3.drop('Place',axis = 1, inplace = True)
        f3 = f3.set_index(pd.to_datetime(f3['Date']))
        f3.drop('Date',axis = 1, inplace = True)

df4 = pd.DataFrame(pd.read_csv('2016.11.csv',sep = ',',encoding='GBK'))
df4.rename(columns={u'时间':'Date',u'地市':'Place',u'流量（Gb）':'DataflowByDay_4G_GB'},inplace = True)
df4 = df4.groupby(df4['Place'])
for name,groupData in df4:
    if name.encode('utf-8') == '镇江':
        f4 = pd.DataFrame(groupData)
        f4.rename(columns={u'时间':'Date',u'流量（GB）':'DataflowByDay_4G_GB'},inplace = True)
        f4.drop('Place',axis = 1, inplace = True)
        f4 = f4.set_index(pd.to_datetime(f4['Date']))
        f4.drop('Date',axis = 1, inplace = True)

df5 = pd.DataFrame(pd.read_csv('2016.12.csv',sep = ',',encoding='GBK'))
df5.rename(columns={u'时间':'Date',u'地市':'Place',u'流量（Gb）':'DataflowByDay_4G_GB'},inplace = True)
df5 = df5.groupby(df5['Place'])
for name,groupData in df5:
    if name.encode('utf-8') == '镇江':
        f5 = pd.DataFrame(groupData)
        f5.rename(columns={u'时间':'Date',u'流量（GB）':'DataflowByDay_4G_GB'},inplace = True)
        f5.drop('Place',axis = 1, inplace = True)
        f5 = f5.set_index(pd.to_datetime(f5['Date']))
        f5.drop('Date',axis = 1, inplace = True)
        f6 = pd.concat([f1,f2,f3,f4,f5],ignore_index= False)
        f6 = f6.sort_index()
f6
```

```python
#对17年1月到12月数据进行清理合并
df6 = pd.DataFrame(pd.read_csv('20171215104003644.csv',sep = ',',encoding='GBK',usecols=[0,1,7]))
df6.rename(columns={u'时间':'Date',u'地市名称':'Place',u'LTE总流量(Gb)':'DataflowByDay_4G_GB'},inplace = True)
df7 = df6.groupby('Place')
for name,groupData in df7:
    if name.encode('utf-8') == '镇江':
        df8 = pd.DataFrame(groupData)
        df8.drop('Place',axis = 1, inplace = True)
        df9 = df8.set_index(pd.to_datetime(df8['Date']))
        df9.drop('Date',axis = 1, inplace = True)
        df10 = df9.sort_index()
        df11 = df10.resample('D',how='sum')
        df12 = df11[df11.index<'2017-12-15']
df13 = pd.concat([f6,df12],ignore_index= False)
df13
```

```python
df13.loc[df13.index =='2016-11-10','DataflowByDay_4G_GB'].values[0] = (df13['2016-11-09':'2016-11-11'].sum()/2).values[0]
df13.to_csv(u'C:\Jupyter WorkSpace\JiangSu Dataflow\lstm\Data Set\Arima预测\oridata13.csv',encoding='utf-8')
```

```python
#整理lstm训练数据
df14 = df13.groupby(pd.TimeGrouper("M"))
y = []
x = 0
for i,j in list(df14):
    if x <len(df14):
        if '04-30' in str(i) or '06-30' in str(i) or '04-30' in str(i) or '09-30' in str(i) or '11-30' in str(i): 
            df6 = pd.DataFrame(list(df14)[x][1])
            df7 = df6.iloc[-1:,[0]]
            y.append(df6.append(df7))
        elif '02-28' in str(i):
            df6 = pd.DataFrame(list(df14)[x][1])
            df7 = df6.iloc[-1:,[0]]
            df8 = df6.append(df7)
            df9 = df8.append(df7)
            df10 = df9.append(df7)
            y.append(df10)
        else:
            y.append(list(df14)[x][1])
        x = x + 1
    else:
        break
df15 = pd.concat(y,ignore_index=False)  
# df15.to_excel('data00.xlsx',encoding='utf-8')
df15
```
