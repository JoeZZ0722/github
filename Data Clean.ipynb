```python
f9 = pd.DataFrame(pd.read_excel(u'国庆节数据.xlsx',sep = '|',encoding='utf-8'))
df1 = pd.DataFrame(pd.read_excel(u'16.39.xlsx',sep = '|',encoding='utf-8'))
pd.set_option('precision', 7)
f10 = f9.groupby(u'地市')
df2 = df1.groupby(u'地市')
df2
```

```python
for name,groupData in f10:
    if name.encode('utf-8') == '徐州':
        f5 = pd.DataFrame(groupData)
        f5.rename(columns={u'时间':'Date',u'流量（GB）':'DataflowByDay_4G_GB'},inplace = True)
        f5.drop(u'地市',axis = 1, inplace = True)
        f6 = f5.set_index(pd.to_datetime(f5['Date']))
        f6.drop('Date',axis = 1, inplace = True)
        f7 = pd.concat([f4,f6],ignore_index= False)
        f8 = f7.sort_index()
        dataflow = f8['DataflowByDay_4G_GB']
        data_mean=dataflow.mean()
        data_std=dataflow.std()
#         f11=f8[f8['DataflowByDay_4G_GB']>=(data_mean-1.96*data_std)]
for name1,groupData1 in df2:
    if name1.encode('utf-8') == '徐州':
        df5 = pd.DataFrame(groupData1)
        df5.rename(columns={u'时间':'Date',u'流量（Gb）':'DataflowByDay_4G_GB'},inplace = True)
        df5.drop(u'地市',axis = 1, inplace = True)
        df5['time'] = df5['Date'].apply(lambda x:str(x.encode('utf-8')).split()[0])
        df5.index = pd.to_datetime(df5['time'])
        df5.drop({'Date','time'},axis = 1, inplace = True)
        df5.index.rename('Date',inplace = True)
        df5['DataflowByDay_4G_GB'] = df5['DataflowByDay_4G_GB']/8
        df7 = pd.concat([f8,df5],ignore_index= False)        
        df8 = df7.sort_index()
        dataflow = df8['DataflowByDay_4G_GB']
        data_mean=dataflow.mean()
        data_std=dataflow.std()
#         df10=df8[df8['DataflowByDay_4G_GB']>=(data_mean-1.96*data_std)]
        pd.set_option('precision', 7)
        df8.to_csv(u'data010.csv',encoding='utf-8')
```
