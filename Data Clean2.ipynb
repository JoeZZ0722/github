```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# 对11地市的全天量数据进行清洗合并（不包含镇江和扬州）
pd.set_option('precision', 7)
df1 = pd.DataFrame(pd.read_csv('dataflow_011.csv',sep = ',',encoding='utf-8'))
# df1
```

```python
# df1[df1.columns[1]]
```

```python
df1['ts']=pd.to_datetime(df1['Data'],format=u'%Y\u5e74%m\u6708%d\u65e5')
df1 = df1.drop(['Data'],axis = 1)
df2=df1.set_index(df1['ts'])
df2.index.rename('Date',inplace = True)
df3=df2.drop(df2.columns[-1],axis = 1)
df4 = df3.sort_index()
df4 = df4[df4.index<'2017-09-01']
# df4
```

```python
df5 = pd.DataFrame(pd.read_csv('20171215104003644.csv',sep = ',',encoding='GBK',usecols=[0,1,7]))
df5.rename(columns={u'时间':'Date',u'地市名称':'Place',u'LTE总流量(Gb)':'DataflowByDay_4G_GB'},inplace = True)
df6 = df5.groupby('Place')
for name,groupData in df6:
    if name.encode('utf-8') == '盐城':
        df7 = pd.DataFrame(groupData)
        df7.drop('Place',axis = 1, inplace = True)
        df8 = df7.set_index(pd.to_datetime(df7['Date']))
        df8.drop('Date',axis = 1, inplace = True)
        df9 = df8.sort_index()
        df10 = df9.resample('D',how='sum')
        df11 = df10[df10.index>'2017-08-31']
        df12 = df11[df11.index<'2017-12-01']
# df12
```

```python
df13 = pd.concat([df4,df12],ignore_index= False)
# df13
```

```python
df14 = df13.groupby(pd.TimeGrouper("M"))
y = []
x = 0
for i,j in list(df14):
    if x <16:
        if '04-30' in str(i) or '06-30' in str(i) or '04-30' in str(i) or '09-30' in str(i) or '11-30' in str(i): 
            df6 = pd.DataFrame(list(df14)[x][1])
            df7 = df6.iloc[-1:,[0]]
            y.append(df6.append(df7))
        elif '02-28' in str(i):
            df6 = pd.DataFrame(list(df14)[x][1])
            df7 = df6.iloc[-1:,[0]]
            df8 = df6.append(df7)
            df9 = df8.append(df7)
            df10 = df9.append(df7)
            y.append(df10)
        else:
            y.append(list(df14)[x][1])
        x = x + 1
    else:
        break
df15 = pd.concat(y,ignore_index=False)  
df15.to_excel('data11.xlsx',encoding='utf-8')
```
